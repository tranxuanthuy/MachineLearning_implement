```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
```

# load_data from sklearn

```python
iris = datasets.load_iris()
```

```python
# features selection 2, 3 for visualization in 2D
X = iris.data[:, [2, 3]]
y = iris.target
```

```python
for target in np.unique(y):
    plt.scatter(X[y == target, 0], X[y == target, 1])
    
```

# Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                   random_state=1,
                                                   stratify=y)
```

```python
'label count in y:', np.bincount(y)
```

```python
'label count in y train, y test: ', np.bincount(y_train), np.bincount(y_test)
```

# preprocessing

## features scaling 

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
sc.fit(X_train)
```

```python
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

# training model

```python
from sklearn.linear_model import Perceptron
```

```python
perceptron = Perceptron(eta0=0.1,
                        random_state=1)
```

```python
perceptron.fit(X_train_std, y_train)
```

# Evaluate

## Classification accuracy

```python
y_pred = perceptron.predict(X_test_std)
'accuracy score: ', (y_pred == y_test).sum() / y_test.shape[0]
```

```python
'Missclassified examples: ', (y_pred != y_test).sum()
```

```python
from sklearn.metrics import accuracy_score
```

```python
'Accuracy score from sklearn metrics: ', accuracy_score(y_test, y_pred)
```

## classification error

```python
'classification error: ', round((y_pred != y_test).sum() / y_test.shape[0], 5)
```

# visualization

```python
from matplotlib.colors import ListedColormap
def plot_desicion_boundary(X, y, classifier, test_idx):
    colors = ('red', 'blue', 'green', 'black')
    cmap = ListedColormap(colors[0:len(np.unique(y))])
    markers = ['o', 's', '^', '<']
    resolution = 0.02
    x1_min, x1_max = X[:, 0].min(), X[:, 0].max()
    x2_min, x2_max = X[:, 1].min(), X[:, 1].max()
    xx1, xx2 = np.meshgrid(np.arange(x1_min - 1, x1_max + 1, resolution),
                            np.arange(x2_min - 1, x2_max + 1, resolution))
    labels = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    labels = labels.reshape(xx1.shape)
    plt.contourf(xx1, xx2, labels, cmap=cmap, alpha=0.2)

    for idx, class_ in enumerate(np.unique(y)):
        plt.scatter(X[y == class_, 0], X[y == class_, 1], color = colors[idx],
                    marker = markers[idx],
                   edgecolor='black')
    if test_idx:
        plt.scatter(X[test_idx, 0], X[test_idx, 1],
                    color='none',
                    edgecolor='black',
                    alpha=1.0,
                    linewidth=1,
                   marker='o',
                    s=100,
                    label='test set'
                   )
```

```python
X_combined = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))
```

```python
plot_desicion_boundary(X_combined, y_combined, perceptron,
                      test_idx=range(100, 150))
```
